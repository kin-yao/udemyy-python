## Merging DataFrames

```python
import pandas as pd
import numpy as np
```

```python

df1 = pd.DataFrame({
    'Employees': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'Department': ['HR', 'IT', 'Finance', 'Marketing'],
    'Year': [2019, 2020, 2021, 2022]
})

df2 = pd.DataFrame({
    'Employees': ['Bob', 'Alice', 'Diana', 'Charlie'],
    'Country': ['USA', 'UK', 'Canada', 'Australia'],
    'Salary': [70000, 80000, 75000, 90000]
})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, how= "right")
```

### saving dataframes with csv and xlsx extension


```python
df3 = pd.merge(df1, df2, how= "right")
```

```python
df3.to_csv("df3.csv")
```

```python
df4 = pd.merge(df1, df2, how = "left")
```

```python
df4.to_excel("df4.xlsx")
```

## Seaborn library for figure aesthetics

```python
##lets start with the normal matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure()
x1 = [10, 20, 5, 40 ,8]
x2 = [30, 43, 9, 7, 20]
plt.plot(x1, label = 'Group A')
plt.plot(x2, label = 'Group B')
plt.legend()
plt.show()
```

```python
# now moving on to a visualization with seaborn>>>
import  seaborn as sns
sns.set()
plt.figure()

x1 = [10, 20, 5, 40 ,8]
x2 = [30, 43, 9, 7, 20]
plt.plot(x1, label = 'Group A')
plt.plot(x2, label = 'Group B')
plt.legend()
plt.show()
```

- As we can see seaborn has added two topping to the above visualization:
1. Aesthetics of the plot.
2. Scales elements of the visualization so that it can be used in different contexts.

### set_style()

```python
sns.set_style("whitegrid")
plt.figure()

x1 = [10, 20, 5, 40 ,8]
x2 = [30, 43, 9, 7, 20]
plt.plot(x1, label = 'Group A')
plt.plot(x2, label = 'Group B')
plt.legend()
plt.show()
```

```python
sns.set_style("darkgrid")
plt.figure()

x1 = [10, 20, 5, 40 ,8]
x2 = [30, 43, 9, 7, 20]
plt.plot(x1, label = 'Group A')
plt.plot(x2, label = 'Group B')
plt.legend()
plt.show()
```

### set_context()

```python
sns.set_context("poster")
plt.figure()

x1 = [10, 20, 5, 40 ,8]
x2 = [30, 43, 9, 7, 20]
plt.plot(x1, label = 'Group A')
plt.plot(x2, label = 'Group B')
plt.legend()
plt.show()
```

## Squarify Library

```python
pip install squarify
```

- The squarify library in Python is used for creating treemaps, which are a type of visualization technique that displays hierarchical data as a set of nested rectangles. Treemaps are particularly useful for visualizing hierarchical structures such as the composition of a dataset or the hierarchical clustering of items.

- The main purpose of the squarify library is to generate these treemaps efficiently and effectively. It provides functions to layout and visualize hierarchical data in a way that emphasizes the relative sizes of the different elements while also maintaining a compact and aesthetically pleasing layout.

```python
import matplotlib.pyplot as plt
import squarify as s
import seaborn as sns

colors = sns.light_palette("Brown",4)
s.plot(sizes = [50, 25, 10, 15], label = ["Group A", "Group B", "Group C", "Group D"], color = colors)
plt.show()
```
