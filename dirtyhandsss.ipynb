## Merging DataFrames

```python
import pandas as pd
import numpy as np
```

```python

df1 = pd.DataFrame({
    'Employees': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'Department': ['HR', 'IT', 'Finance', 'Marketing'],
    'Year': [2019, 2020, 2021, 2022]
})

df2 = pd.DataFrame({
    'Employees': ['Bob', 'Alice', 'Diana', 'Charlie'],
    'Country': ['USA', 'UK', 'Canada', 'Australia'],
    'Salary': [70000, 80000, 75000, 90000]
})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, how= "right")
```

### saving dataframes with csv and xlsx extension


```python
df3 = pd.merge(df1, df2, how= "right")
```

```python
df3.to_csv("df3.csv")
```

```python
df4 = pd.merge(df1, df2, how = "left")
```

```python
df4.to_excel("df4.xlsx")
```

## Seaborn library for figure aesthetics

```python
##lets start with the normal matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure()
x1 = [10, 20, 5, 40 ,8]
x2 = [30, 43, 9, 7, 20]
plt.plot(x1, label = 'Group A')
plt.plot(x2, label = 'Group B')
plt.legend()
plt.show()
```

```python
# now moving on to a visualization with seaborn>>>
import  seaborn as sns
sns.set()
plt.figure()

x1 = [10, 20, 5, 40 ,8]
x2 = [30, 43, 9, 7, 20]
plt.plot(x1, label = 'Group A')
plt.plot(x2, label = 'Group B')
plt.legend()
plt.show()
```

- As we can see seaborn has added two topping to the above visualization:
1. Aesthetics of the plot.
2. Scales elements of the visualization so that it can be used in different contexts.

### set_style()

```python
sns.set_style("whitegrid")
plt.figure()

x1 = [10, 20, 5, 40 ,8]
x2 = [30, 43, 9, 7, 20]
plt.plot(x1, label = 'Group A')
plt.plot(x2, label = 'Group B')
plt.legend()
plt.show()
```

```python
sns.set_style("darkgrid")
plt.figure()

x1 = [10, 20, 5, 40 ,8]
x2 = [30, 43, 9, 7, 20]
plt.plot(x1, label = 'Group A')
plt.plot(x2, label = 'Group B')
plt.legend()
plt.show()
```

### set_context()

```python
sns.set_context("poster")
plt.figure()

x1 = [10, 20, 5, 40 ,8]
x2 = [30, 43, 9, 7, 20]
plt.plot(x1, label = 'Group A')
plt.plot(x2, label = 'Group B')
plt.legend()
plt.show()
```

## Popular Seaborn plots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

example_data = pd.read_csv("Salary_Data.csv")
example_data
```

```python
sns.set(style = "whitegrid")
sns.barplot(x = "Education Level", y = "Salary", hue="Gender", data=example_data)
```

```python
sns.barplot(x = "Education Level", y = "Salary", data = example_data )
```

```python
sns.distplot(example_data["Age"])
```

```python
sns.kdeplot(example_data["Age"])
```

```python
sns.set(style = "whitegrid")
sns.jointplot(x = "Salary", y = "Age", data = example_data)
```

```python
sns.set(style = "ticks", color_codes = True)
g=sns.pairplot(example_data, hue = "Groups")
```

```python
sns.set(style = "whitegrid")
sns.violinplot("Education Level", "Salary", hue = "Gender", data = example_data, split = True, cut = 0 )
```
