```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(1, 9)
y = np.arange(3, 27, 3)

plt.xlabel("x_axis")
plt.ylabel("y_label")
plt.title("sample plot")
plt.plot(x, y)
plt.show()
```

```python
plt.barh(x, y)
plt.step(x, y)
plt.polar(x,y)
plt.bar(x, y)
```

## Merging DataFrames

```python
import pandas as pd
import numpy as np
```

```python

df1 = pd.DataFrame({
    'Employees': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'Department': ['HR', 'IT', 'Finance', 'Marketing'],
    'Year': [2019, 2020, 2021, 2022]
})

df2 = pd.DataFrame({
    'Employees': ['Bob', 'Alice', 'Diana', 'Charlie'],
    'Country': ['USA', 'UK', 'Canada', 'Australia'],
    'Salary': [70000, 80000, 75000, 90000]
})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, how= "right")
```

### saving dataframes with csv and xlsx extension


```python
df3 = pd.merge(df1, df2, how= "right")
```

```python
df3.to_csv("df3.csv")
```

```python
df4 = pd.merge(df1, df2, how = "left")
```

```python
df4.to_excel("df4.xlsx")
```
